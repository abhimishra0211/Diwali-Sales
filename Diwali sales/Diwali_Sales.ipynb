# Exploratory Data Analysis


# Gender

```python
df['Gender'].value_counts()
```

```python
df['Gender'].value_counts().plot(kind='pie',autopct='%0.2f%%',
                                explode=[0,0.1])
plt.title('Gender Count')
plt.show()
```

Above graph shows Female is 7842 and male is 3409.

```python
ab = df.groupby(['Gender'], as_index = False) ['Amount'].sum().sort_values(by = 'Amount', ascending = False)
ab
```

```python
plt.figure(figsize=(5,3))
ac=sns.barplot(x ='Gender', y='Amount',data =ab)
for a,b in ab.iterrows():
    ac.text(a,b['Amount'],b['Amount'], ha='center')
plt.title('Gender Wise Purchasing Amount')    
plt.show()    
```

In above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men.

# Age Group

```python
df['Age Group'].value_counts()
```

```python
plt.figure(figsize=(7,3))
aa=sns.countplot(data=df, x='Age Group', hue='Gender')
for bars in aa.containers:
    aa.bar_label(bars)
plt.grid()    
plt.title('Age Group Count')
plt.show()
```

```python
sa=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sa
```

```python
plt.figure(figsize=(10,3))
q=sns.barplot(x ='Age Group',y='Amount' ,data=sa)
for bars in q.containers:
    q.bar_label(bars)
plt.title('Age Group Vs Amount')
plt.show()
```

In above graphs we can see that most of the buyers are of age group between 26-35 yrs female.

```python
sns.lineplot(x ='Age Group',y='Amount' ,data=sa)
plt.title('Age Group Vs Amount')
plt.show()
```

# State

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
plt.figure(figsize=(15,5))
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
plt.title('State Vs Orders ')
plt.show()
```

```python
sa=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
plt.figure(figsize=(15,4))
sns.barplot(data =sa, x ='State',y='Amount')
plt.title('State Wise Total Amount Spend')
plt.show()
```

In above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

# Marital_Status

```python
ms=sns.countplot(data = df, x = 'Marital_Status')
plt.title('Marital Status Count')
for bars in ms.containers:
    ms.bar_label(bars)
```

```python
ss=df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
ss
```

```python
plt.figure(figsize=(7,3))
w=sns.barplot(data=ss,x='Marital_Status',y='Amount', hue='Gender')
for bars in w.containers:
    w.bar_label(bars)
plt.title('Marital_Status Vs Amount')
plt.show()
```

In above graphs we can see that most of the buyers are married (women) and they have high purchasing power.

# Occupation

```python
plt.figure(figsize=(20,4))
ax = sns.countplot(data = df, x = 'Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Occupation Count')
plt.show()
```

```python
plt.figure(figsize=(15,4))
so=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(data=so, x ='Occupation',y='Amount')
plt.title('Occupation Vs Amount')
plt.show()
```

# Product_Category

```python
plt.figure(figsize=(25,4))
pc=sns.countplot(data=df, x='Product_Category')
plt.title('Product Category Count')
for bars in pc.containers:
    pc.bar_label(bars)
```
